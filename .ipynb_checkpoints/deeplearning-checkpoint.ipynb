#### text generation

```python
def read_file(path):
    with open(path) as f:
        str_text = f.read()
    return str_text
```

```python
#read_file('moby_dick_four_chapters.txt')
#read_file('melville-moby_dick.txt')
```

```python
import spacy
```

```python
nlp = spacy.load('en',disable=['parser','tagger','ner'])
```

```python
nlp.max_length = 1198623
```

```python
def seperate_punctuation(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n']
```

```python
d = read_file('moby_dick_four_chapters.txt')
```

```python
tokens = seperate_punctuation(d)
```

```python
tokens[:10]
```

```python
len(tokens)
```

```python
# 25 words ---> predict 26
```

```python
train_len = 25 + 1

text_sequences = []

for i in range(train_len,len(tokens)):
#     print(i-train_len,i)
#     print('---')
    seq = tokens[i-train_len:i]
    text_sequences.append(seq)
    
```

```python
tokenizer.fit_on_texts(text_sequences)
```

```python
sequences = tokenizer.texts_to_sequences(text_sequences)
```

```python
y=to_categorical(y,num_classes=vocab_size+1)
```

```python
X.shape
```

```python
seq_len = X.shape[1]
seq_len
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding
```

```python
def create_model(vocab_size,seq_len):
    
    model = Sequential()
    model.add(Embedding(vocab_size,seq_len,input_length=seq_len))
    model.add(LSTM(50,activation='tanh',return_sequences=True))
    model.add(LSTM(50))
    model.add(Dense(50,activation='relu'))
    
    model.add(Dense(vocab_size,activation='softmax'))
    
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    model.summary()
    
    return model
```

```python
model=create_model(vocab_size+1,seq_len)
```

```python
from pickle import dump,load
```

```python
model.fit(X,y,batch_size=128,epochs=20,verbose=1)
```

```python
from keras.preprocessing.sequence import pad_sequences,sequence
```

```python
def generate_text(model,tokenizer,seq_len,seed_text,num_gen_words):
    
    output_text = []
    
    input_text = seed_text
    
    for i in range(num_gen_words):
        
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        #print(input_text,'---',encoded_text)
        
        pad_encoded = pad_sequences([encoded_text],maxlen=seq_len,truncating='pre')
        #print('pad',pad_encoded)
        
        pred_word_ind = model.predict_classes(pad_encoded,verbose=0)
        #print('##',pred_word_ind)
        
        pred_word_ind = model.predict_classes(pad_encoded,verbose=0)[0]
        #print('$$',pred_word_ind)
        
        pred_word = tokenizer.index_word[pred_word_ind]
        #print('predword',pred_word)
        
        input_text = ' ' + pred_word
        #print('&&',input_text)
        
        output_text.append(pred_word)
        
    
    return ' '.join(output_text)
```
